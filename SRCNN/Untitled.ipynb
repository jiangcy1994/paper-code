```python
import sys
sys.path.append('../utils/')
from compose import *
from functools import partial
from keras.layers import Add, Conv2D, Input, ReLU
from keras.models import Model
```

```python
inp = Input((80,80,3))
```

```python
conv2d = partial(Conv2D, kernel_size=3, strides=1, padding='same', kernel_initializer='he_normal')
```

```python
compose(
    conv2d(64),
    ReLU('relu2')
    conv2d(64),
)
```
